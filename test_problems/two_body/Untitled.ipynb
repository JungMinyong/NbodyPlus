```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
N=1000
particles={"1":np.zeros((6,N)),"2":np.zeros((6,N))}
for i in range(N):
    f=open(f"../../src/output/snapshot_{5*i+1}.txt", "r")
    f.readline();f.readline();
    for j, line in enumerate(f.readlines()):
        quan = line.split()
        particles[f"{j+1}"][0,i] = quan[1]
        particles[f"{j+1}"][1,i] = quan[2]
        particles[f"{j+1}"][2,i] = quan[3]
        particles[f"{j+1}"][3,i] = quan[4]
        particles[f"{j+1}"][4,i] = quan[5]
        particles[f"{j+1}"][5,i] = quan[6]
```

```python
plt.figure(figsize=(10,10))
plt.scatter(particles['1'][0,::20],particles['1'][1,::20], c='b',alpha=0.5)
plt.scatter(particles['2'][0,::20],particles['2'][1,::20], c='r',alpha=0.5)
plt.grid()
```

```python
E=np.zeros((2,N));
for j in range(2):
    for i in range(N):
        for dim in range(3):
            E[j,i] = E[j,i] + particles[f"{j+1}"][dim+3,i]**2
```
