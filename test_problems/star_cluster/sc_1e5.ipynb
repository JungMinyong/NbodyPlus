```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
num_snap = 200
num_ptcl = 1000
data     =np.zeros((num_ptcl,6,num_snap))
for i in range(num_snap):
    f=open(f"../../src/output/snapshot_{i+1}.txt", "r")
    f.readline();f.readline();
    for j, line in enumerate(f.readlines()):
        quan = line.split()
        data[j,0,i] = quan[1]
        data[j,1,i] = quan[2]
        data[j,2,i] = quan[3]
        data[j,3,i] = quan[4]
        data[j,4,i] = quan[5]
        data[j,5,i] = quan[6]
```

```python
plt.figure(figsize=(10,10))
i=0
plt.scatter(data[:,0,i],data[:,1,i], s=0.5,c='b',alpha=1,zorder=1)
i=0
plt.scatter(data[:,0,i],data[:,1,i], s=0.5,c='r',alpha=1,zorder=2)
i=199
plt.scatter(data[:,0,i],data[:,1,i], s=0.5,c='k',alpha=1,zorder=3)
```
